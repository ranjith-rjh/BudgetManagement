## üìç Functions

```python
def get_data_from_query(query, cursor=cursor) : 
    cursor.execute(query)
    
    colnames = [col.name for col in cursor.description]
    rows = cursor.fetchall()

    data = pd.DataFrame(columns=colnames, data=rows)
    return data
```
