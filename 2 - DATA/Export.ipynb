```python
# Create database connection

conn = ''

try :
    conn = psycopg2.connect(
        host = "localhost",
        port = "5432",
        user = "postgres",
        password = "postgres",
        dbname = "BudgetManagement"
    )    

except Exception as e: 
    print(e)

# Connect to Database and get informations
cursor = conn.cursor()
```

```python
data_tables = [
    'PERIODE',
    'TYPE_FLUX',
    'FIXES',
    'CATEGORIE',
    'FLUX',
    'vue_Flux',
    'vue_Flux_MAct',
    'vue_Flux_SAct',
]
```

```python
def export_table(table) :
    query = f'SELECT * FROM {table}'
    cursor.execute(query)

    colnames = [col.name for col in cursor.description]
    rows = cursor.fetchall()

    data = pd.DataFrame(columns=colnames, data=rows)
    data.to_csv(f'./BACKUP/{table}.csv', sep=';', index=False)
    print(f'‚úÖ EXPORTED : {table}')
```
