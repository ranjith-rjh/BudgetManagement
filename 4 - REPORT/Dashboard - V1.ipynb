## üöö Import

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## ‚öôÔ∏è Data process

```python
data = pd.read_csv('..\\2 - DATA\\BACKUP\\VUE_FLUX_MAct.csv', sep=';')
data = data.drop(columns={'Tags'}, axis=1)
data
```

```python
data_entree = data.copy()
data_entree = data_entree[data_entree['Type de flux'] == 'Entr√©e']
```

```python
data_sortie = data.copy()
data_sortie = data_sortie[data_sortie['Type de flux'] == 'Sortie']
data_sortie
```

## üñåÔ∏è Data Visualisation

```python
data_plot = data_entree.copy()
data_plot = data_entree[['Cat√©gorie', 'Montant']]
data_plot = (
    data_plot.groupby('Cat√©gorie')
    .sum()
    .sort_values('Montant', ascending=False)
)
```

```python
fig, ax = plt.subplots()

p_entree = ax.bar(
    data_plot.index,
    data_plot['Montant'],
    color = '#4361ee'
)

ax.bar_label(p_entree, label_type='center', color = 'white')
plt.title('Entr√©e du mois')
plt.show()
```

```python
data_plot = data_sortie.copy()
data_plot = data_sortie[['Cat√©gorie', 'Montant']]
data_plot = (
    data_plot.groupby('Cat√©gorie')
    .sum()
    .sort_values('Montant', ascending=False)
)
```

```python
fig, ax = plt.subplots(figsize=(30, 10))

p_entree = ax.bar(
    data_plot.index,
    data_plot['Montant'],
    color = '#ff006e'
)

ax.bar_label(p_entree, label_type='edge', color = 'black')
plt.title('D√©penses du mois')
plt.show()
```
